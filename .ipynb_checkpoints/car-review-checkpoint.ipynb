DATA PREPROCESSING

```python
import nltk
import re
from nltk.corpus import stopwords

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer



    
# Initializ INGthe WordNet lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
df= df[:2500]
```

```python
import re

corpus = []
for i in range(0, len(df)):
    processed_review = re.sub('[^a-zA-Z]', ' ',str(df['Review'][i]) )
    processed_review = processed_review.lower()
    processed_review = processed_review.split()
    
    
    processed_review =[lemmatizer.lemmatize(word) for word in processed_review if not word in stopwords.words('english')]
    processed_review=' '.join(processed_review)  # Join the lemmatized words back into a single string ##note when you lemittaize word can broken into single words to combine imto list of stings we use this functiom
    corpus.append(processed_review)
```

```python
corpus
```

```python
#By using SentimentIntensityAnalyzer we can give post,neg,neutal values to our text
from nltk.sentiment import SentimentIntensityAnalyzer

from tqdm.notebook import tqdm

sia=SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores('i was not happy with him')   ##neg ,pos,neu
```

```python
df['corpus'] = corpus
```

```python
df 
```

```python
df = df.drop(['Review'],axis=1)
```

```python
df
```

```python
df["corpus"][0:10].apply(lambda x: sia.polarity_scores(x))  ##for
```

```python
df["corpus"][0:10].apply(lambda x: sia.polarity_scores(x)["compound"])
```

```python
df["polarity_score"] = df["corpus"].apply(lambda x: sia.polarity_scores(x)["compound"])   ##adding the polarity scorce to the datasets
df.head()
```

```python
df["sentiment_label"] = df["corpus"].apply(lambda x: "pos" if sia.polarity_scores(x)["compound"] > 0 else "neg" )
df.head(20)
```

SPLITTING TRAIN AND TEST DATASETS

```python
from sklearn.model_selection import train_test_split

# Assuming you have your data in 'X' (features) and 'y' (labels)
X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size=0.3, random_state=42)
```

MODEL BUILDING

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

logreg.fit(X_train,y_train)
```

PREDICT THE DATASETS

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
