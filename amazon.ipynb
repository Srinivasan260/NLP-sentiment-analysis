```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer



    
# Initializ INGthe WordNet lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## TOKENIZE ,Lemmatize and removing all stopwords
import re

corpus = []
for i in range(0, len(df)):
    processed_review = re.sub('[^a-zA-Z]', ' ',str(df['reviewText'][i]) )
    processed_review = processed_review.lower()
    processed_review = processed_review.split()
    
    
    processed_review =[lemmatizer.lemmatize(word) for word in processed_review if not word in stopwords.words('english')]
    processed_review=' '.join(processed_review)  # Join the lemmatized words back into a single string ##note when you lemittaize word can broken into single words to combine imto list of stings we use this functiom
    corpus.append(processed_review)
```
